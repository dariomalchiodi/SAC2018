```python
import matplotlib.pyplot as plt
import numpy as np

from possibilearn.fuzzifiers import *
```

```python
%matplotlib inline

c = CrispFuzzifier()
cf = c.get_fuzzified_membership(3, [1,2,1,2,3,3,3,3,2,4,7,3,2], lambda x: x)
x = np.arange(0, 10, .01)
plt.plot(x, map(cf, x))
plt.ylim(0, 1.1)
```

```python
def get_fuzzifier_graph(fuzzifier, x):
    memb = fuzzifier.get_fuzzified_membership(3, [1,2,1,2,3,3,3,3,2,4,7,3,2], lambda x: x)
    plt.plot(x, map(memb, x))
    plt.ylim(0, 1.1)
    plt.show()

x = np.arange(0, 10, .01)
```

```python
get_fuzzifier_graph(CrispFuzzifier(), x)
```

```python
get_fuzzifier_graph(LinearFuzzifier(), x)
```

```python
get_fuzzifier_graph(QuantileConstantPiecewiseFuzzifier(), x)
```

```python
get_fuzzifier_graph(QuantileLinearPiecewiseFuzzifier(), x)
```

```python
[get_fuzzifier_graph(ExponentialFuzzifier(alpha), x) for alpha in (.001, 0.07, 0.2, 0.5)]
```
